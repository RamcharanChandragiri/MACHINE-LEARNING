Assgnment-02

Objective:
To practice selecting columns and filtering rows using Pandas.

Tasks:
1. Select score-related columns.
2. Filter students scoring above 70 in math.
3. Filter data based on gender.
4. Count number of students in each category.

Conclusion:
Filtering operations help extract meaningful information from data.



```python
import pandas as pd
df = pd.read_csv('Student_Performance.csv')
score_columns = df.filter(like='score')
print(score_columns)
math_above_70 = df[df['math_score'] > 70]
print(math_above_70)
gender_filtered = df[df['gender'] == 'female']
print(gender_filtered)
category_counts = df['parent_education'].value_counts()
print(category_counts)
```

Assignment-03

Objective:
To perform numerical computations on student scores using NumPy.

Tasks:
1. Convert score columns into NumPy arrays.
2. Compute mean, median, standard deviation.
3. Find minimum and maximum scores.

Conclusion:
NumPy enables efficient numerical analysis of large datasets.


```python
import pandas as pd
import numpy as np
df = pd.read_csv('Student_Performance.csv')
score_columns = df.filter(like='score')
score_array = score_columns.to_numpy()
print(score_array)
mean_scores = np.mean(score_array, axis=0)
median_scores = np.median(score_array, axis=0)
std_scores = np.std(score_array, axis=0)
min_scores = np.min(score_array, axis=0)
max_scores = np.max(score_array, axis=0)
print(mean_scores)
print(median_scores)
print(std_scores)
print(min_scores)
print(max_scores)
```

Assignment-04

Objective:
To visualize student performance using Matplotlib.

Tasks:
1. Plot histogram of math scores.
2. Plot bar chart of average scores.
3. Create scatter plot between math and writing scores.

Conclusion:
Visualization helps understand data distribution and relationships.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('Student_Performance.csv')
math_scores = df['math_score']
plt.hist(math_scores, bins=10, edgecolor='black')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.title('Histogram of Math Scores')
plt.show()
average_scores = df[['math_score', 'science_score', 'english_score']].mean()
plt.bar(['Math', 'science', 'english'], average_scores)
plt.xlabel('Subject')
plt.ylabel('Average Score')
plt.title('Bar Chart of Average Scores')
plt.show()
math_scores = df['math_score']
english_scores = df['english_score']
plt.scatter(math_scores, english_scores)
plt.xlabel('math_Score')
plt.ylabel('English_Score')
plt.title('Scatter Plot of Math vs. Writing Scores')
plt.show()
```

Assignment-05

Objective:
To analyze student performance based on different groups.

Tasks:
1. Group scores by gender.
2. Group scores by parental education.
3. Visualize group-wise average scores using bar charts.

Conclusion:
Group-based analysis reveals performance trends among different categories.


```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('Student_Performance.csv')
gender_grouped = df.groupby('gender')
gender_average = gender_grouped[['math_score', 'science_score', 'english_score']].mean()
gender_average.plot(kind='bar')
plt.xlabel('Gender')
plt.ylabel('Average Score')
plt.title('Group-wise Average Scores by Gender')
plt.show()
parental_education_grouped = df.groupby('parent_education')
parental_education_average = parental_education_grouped[['math_score', 'science_score', 'english_score']].mean()
parental_education_average.plot(kind='bar')
plt.xlabel('Parental Education')
plt.ylabel('Average Score')
plt.title('Group-wise Average Scores by Parental Education')
plt.show()
```
