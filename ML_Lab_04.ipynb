**SUBTASK-01**

Objective:
To load and understand the salary dataset for linear regression using Scikit-learn.

Tasks:
1. Download the dataset from Kaggle.
2. Load the dataset using Pandas.
3. Display first five rows.
4. Identify input and output variables.

Conclusion: The dataset was loaded successfully and variables were identified.


```python
#load the dataset using pandas
import pandas as pd

df = pd.read_csv('/content/Salary_dataset.csv')
```

```python
#display first five rows
df.head()
```

```python
#identify input and output variables
x = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

**SUBTASK-02**
Objective:
To implement Linear Regression using Scikit-learn library.

Tasks:
1. Split data into training and testing sets.
2. Import LinearRegression from sklearn.
3. Train the model using fit().
4. Predict output values.

Conclusion: Linear regression model was successfully implemented using Scikit-learn.


```python
#1.Split data into training and testing sets.
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
#2 import linear regression from sklearn
from sklearn.linear_model import LinearRegression
```

```python
#3 train the model using fit()
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
#4 predict output values
y_pred = regressor.predict(x_test)
#
```

**SUBTASK-03**
Objective:
To evaluate the performance of Scikit-learn Linear Regression model.

Tasks:
1. Calculate Mean Squared Error.
2. Calculate R-squared value.
3. Interpret model accuracy.

Conclusion: Model performance metrics were evaluated successfully.


```python
# calculate mean squared error
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python
#2.calculate R-squared value
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print("R-squared Value:", r2)
```

```python
#3.interpret model accuracy
if r2 > 0.7:
    print("The model is accurate.")
else:
    print("The model is not accurate.")
```

**SUBTASK-04**

Tasks:

```python
# compare slope and intercept values
print("Slope:", regressor.coef_)
print("Intercept:", regressor.intercept_)
```

```python
#2 compare prediction results and visualise them
import matplotlib.pyplot as plt

# Assuming YearsExperience is the second column (index 1) in x_test
plt.scatter(x_test[:, 1], y_test, color='red', label='Actual Salaries')
plt.plot(x_test[:, 1], y_pred, color='blue', label='Predicted Salaries')
plt.title('Salary vs Years of Experience (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
#3 compare error values
print("Mean Squared Error:", mse)
print("R-squared Value:", r2)
```
